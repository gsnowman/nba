```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
fig = plt.figure()

import sqlite3
conn = sqlite3.connect('nba.sqlite')
c = conn.cursor()
```

```python
last = 'Cousins'

for row in c.execute("SELECT name, yahoo from player_ids where name like '%%%s%%';" % last):
    print row
```

```python
def get_years_ago(y):
    return -1 * (2013 - int(y.split('-')[0]))

def get_name(player_id):
    for row in c.execute("SELECT first || ' ' || last from players where player_id = %d;" % player_id):
        return row[0]

def get_years_values(player_id):
    years = []; values = []
    for row in c.execute("SELECT season, total from season_values where player_id = %d order by season asc;" % player_id):
        years.append(get_years_ago(row[0]))
        values.append(row[1])
    return (years, values)


fig = plt.figure(figsize=(8,6))
ax = plt.subplot(111)
for p in (3704, 4563, 4244, 4612, 3930, 5007, 4390, 3706, 4720 ):
    data = get_years_values(p)
    name = get_name(p)
    ax.plot(data[0], data[1], label=name)


ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=3)
plt.show()

    
```
